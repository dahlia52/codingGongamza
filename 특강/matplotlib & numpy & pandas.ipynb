## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([30,20,50,10])
plt.show()
```

```python
plt.plot([2,4,6,8], [10,30,20,60])
plt.show()
```

```python
plt.bar(['A','B','C','D'], [30,10,20,70])
plt.show()
```

```python
plt.pie([10,30,10,50], labels = ['A','B','C','D'])
plt.show()
```

```python
# 그래프 옵션 지정
plt.plot([2,4,6,8], [10,30,20,60], color = 'skyblue', linestyle = '--', label = '1st')
plt.plot([1,3,5,7], [30,40,10,50], color = 'r', ls = ':', label = '2nd')
plt.title('plot graph')
plt.legend()
plt.show()
```

```python
# 그래프 옵션 지정
plt.plot([2,4,6,8], [10,30,20,60], 'r.', label = '1st')
plt.plot([1,3,5,7], [30,40,10,50], 'g^', label = '2nd')
plt.title('plot graph')
plt.legend()
plt.show()
```

## Numpy

```python
import numpy as np
```

```python
a = np.array([1,2,3,4,5])
b = np.array([[1,2,3,4],
              [10, 20, 30, 40]])

print(a)
print()
print(b)
```

```python
a = np.arange(5)
b = np.arange(10, dtype=float)
c = np.arange(3.1, 5, 0.25)

print(a)
print()
print(b)
print()
print(c)
```

```python
a = np.arange(10)

print(a[5])
print(a[:])
print(a[:4])
print(a[:-1])
print(a[3::2])
print(a[7:2:-1])
print(a[::-1])
```

```python
a = np.arange(20).reshape((5, 4))

print(a)
print(a[0:5, 1])
print(a[:,1])
print(a[1:3, :])

print(a[-1])
print(a[:,-1])
```

## Pandas

```python
import pandas as pd
```

```python
score = {'Names':['Kim','Park','Kim','Lee','Lee'],
         'Year':[2019, 2020, 2021, 2022, 2023],
         'Points':[1.5, 1.8, 2.5, 6.1, 0.1]}
score_df = pd.DataFrame(score)
score_df
```

```python
dates = pd.date_range('20230707', periods=6)
print(dates)
print()

date_df = pd.DataFrame(np.random.choice(100, (6,3)),
                       index=dates, columns=list('ABC'))
date_df
```
